```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Fancy Plot 1: Scatter Plot with Bubble Size Variation
plt.figure(figsize=(8,6))
# Bubble sizes vary based on sin function
bubble_sizes = np.abs(np.sin(x)* 1000) 
plt.scatter(x,y1,s=bubble_sizes,c='blue',alpha=0.5,label='sin(x)')
plt.scatter(x,y2,s=bubble_sizes,c='red', alpha=0.5, label='cos(x)')
plt.title('Scatter Plot with Bubble Size Variation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fancy Plot 2: 3D Surface Plot
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(x,x)
Z = np.sin(X) * np.cos(Y)
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Surface Plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.show()
```

```python
# Fancy Plot 3: Violin Plot
plt.figure(figsize=(8,6))
data = [np.random.normal(0, std, 100) for std in range(1,4)]
plt.violinplot(data,showmeans=False,showmedians=True)
plt.title('Violin Plot')
plt.xlabel('Data Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Fancy Plot 4: Heatmap
plt.figure(figsize=(8,6))
data = np.random.rand(10,10)
plt.imshow(data,cmap='hot',interpolation='nearest')
plt.title('HeatMap')
plt.colorbar()
plt.show()
```

```python
# Fancy Plot 5: Histogram with Multiple Distributions
plt.figure(figsize=(8,6))
data1 = np.random.normal(0,1,1000)
data2 = np.random.normal(2,1.5,1000)

plt.hist([data1,data2],bins=30,alpha=0.5,label=['Data 1','Date 2'])
plt.title('Histogram with Multiple Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fancy Plot 6: Line Plot with Error Bars
plt.figure(figsize=(8,6))
plt.errorbar(x,y1,yerr=0.1,fmt='-o',color='green',label='sin(x)')
plt.errorbar(x,y2,yerr=0.1,fmt='-o',color='purple',label='cos(x)')
plt.title('Line Plot with Error Bars')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fancy Plot 7: Box Plot
plt.figure(figsize=(8,6))
plt.boxplot(data,vert=False,patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Fancy Plot 8: Area Plot
plt.figure(figsize=(8, 6))
plt.fill_between(x,y1,color='orange',alpha=0.5,label='sin(x)')
plt.fill_between(x,y2,color='pink',alpha=0.5,label='cos(x)')
plt.title('Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fancy Plot 9: Pie Chart
plt.figure(figsize=(8,6))
labels = ['A','B','C','D']
sizes = [15,30,45,10]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
# Fancy Plot 10: Bar Plot with Horizontal Lines
plt.figure(figsize=(8, 6))
plt.bar(x,y1,color='blue', label='sin(x)')
plt.bar(x, y2, color='red', alpha=0.5, label='cos(x)')
plt.axhline(0, color='black', linestyle='--')
plt.title('Bar Plot with Horizontal Lines')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
